```python
import math
def y1(x):
    y=math.sqrt(math.sin(x)+1)
    return y
y1(0)
```

```python
def y2(x):
    y=math.sqrt((math.sin(x))/(math.cos(x)))
    return y
y2(1)
```

```python
def y3(x):
    y=math.sqrt(math.tan(x)/math.sin(x))
    return y
y3(1)
```

```python
def y4(x):
    y=math.sqrt(1/1+(math.exp(x)))
    return y
y4(1)
```

```python
def y5(w,a):
    y=w-(a*(math.sqrt(w)/math.sqrt(a)))
    return y 
y5(4,4)
```
